```python
%matplotlib inline

# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

states_df = pd.read_csv('PlottingData2.csv')

states_df.tail()
```

```python
x_axis = np.arange(len(states_df))
tick_locations = [value+0.4 for value in x_axis]

plt.figure(figsize=(20,3))
plt.bar(x_axis, states_df["Per Capita Personal Income"], color='b', alpha=0.3, align="edge")
plt.xticks(tick_locations, states_df["State"], rotation=70)

plt.xlim(-0.50, len(x_axis))
plt.ylim(0, max(states_df["Per Capita Personal Income"])+10000)

plt.title("Per Capita Personal Income")
plt.xlabel("State")
plt.ylabel("Per Capita Personal Income")

plt.savefig("Per_Capita_Income.png")
plt.show()
```
