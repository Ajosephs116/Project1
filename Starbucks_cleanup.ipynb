```python
# Import Dependencies
import pandas as pd
```

```python
csv_path = "Data/starbucks_clean.csv"
starbucks_data = pd.read_csv(csv_path)
starbucks_data.head()
```

```python
us_starbucks_data = starbucks_data.loc[starbucks_data["Country"] == "US", :]
us_starbucks_data.head()
```

```python
df_us_starbucks_data = us_starbucks_data.set_index("Postcode")
df_us_starbucks_data.head()
```

```python
value_counts_zip = df_us_starbucks_data["Postcode"].value_counts()
#value_counts_zip
df_counts = pd.DataFrame(value_counts_zip)
df_counts
```

```python
states_starbucks_data = us_starbucks_data.groupby(["State/Province"])
#print(states_starbucks_data)
states_starbucks_data.count().head()
```
