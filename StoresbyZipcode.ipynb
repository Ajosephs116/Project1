```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

total_df = pd.read_csv('Data/ProjectOne.csv')

total_df.head()
```

```python
sort_total_df = total_df.sort_values(by=['total'], ascending=False)
top_total_df = sort_total_df.head(20)
top_total_df_100 = sort_total_df.head(100)
top_total_df
```

```python
x_axis = np.arange(len(top_total_df))
tick_locations = [value+0.4 for value in x_axis]

plt.figure(figsize=(20,3))
plt.bar(x_axis, top_total_df["total"], color='b', alpha=0.3, align="edge")
plt.xticks(tick_locations, top_total_df["zipcode"], rotation=70)

plt.xlim(-0.50, len(x_axis))
plt.ylim(0, max(top_total_df["total"])+5)

plt.title("Total Number of Stores by Zipcode")
plt.xlabel("Zipcode")
plt.ylabel("Total Number of Stores")

plt.figtext(0.65,0.8, "Top 3 States: Nevada(89109), Ohio(45069), Kentucky(41042)")

plt.savefig("StoresbyZipcode.png")
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1.0,1.75])

ax.set_title("House Price")
ax.set_xlabel("Zipcode")
ax.set_ylabel("Number of Stores")
ax.grid(True)

x = top_total_df_100["zipcode"]
y = top_total_df_100["total"]
s = top_total_df_100["2018-12"]

plt.scatter(x,y,s=0.001*s,
           marker ='o', facecolors ='red', edgecolors = 'black', alpha=0.8)


plt.savefig("HousePrice.png")
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1.0,1.75])

ax.set_title("House Price")
ax.set_xlabel("State")
ax.set_ylabel("Number of Stores")
ax.grid(True)
plt.xticks(tick_locations, top_total_df["StateName"], rotation=70)



x = top_total_df["StateName"]
y = top_total_df["total"]
s = top_total_df["2018-12"]

plt.scatter(x,y,s=0.001*s,
           marker ='o', facecolors ='coral', edgecolors = 'black', alpha=0.8)


plt.savefig("HousePrice.png")
plt.show()
```
